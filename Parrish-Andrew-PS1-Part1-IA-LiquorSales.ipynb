```python
#imports
import pandas as pd
import os
import sys
#sets rows to all be visible
pd.set_option("display.min_rows", 201)
pd.set_option("display.max_rows", 201)
# load_dataset function
def load_dataset(file):
    """ takes in the .csv file and creates a dataframe"""
    try:
        df = pd.read_csv(file, delimiter=",", low_memory = False)
        return df
    #handling if the file isn't found
    except FileNotFoundError:
        print("Cannot find the file!")
        sys.exit()
#assigning the file name
file = os.getcwd() + "\\Iowa_Liquor_Sales.csv"
#function call and dataframe assignment
dataframe = load_dataset(file)
#dataset_info function
def dataset_info(dataframe):
    """ takes in the dataframe and returns the amount of rows and columns, as well as the individual column names and the printing of the answers to the questions"""
    #gets the shape of the dataframe
    rowsandcolumns = dataframe.shape
    #assigns the elements of the shape into variables for printing
    rows = rowsandcolumns[0]
    columns = rowsandcolumns[1]
    print("The data has " + str(rows) + " rows and " + str(columns) + " columns.")
    #prints every column in the dataframe
    for col in dataframe.columns:
        print(col)
    #printing out the first 15 rows of the dataframe
    dataframe.head(15)
    #converts all of the county names to uppercase for grouping
    dataframe['County'] = dataframe['County'].str.upper()
    #takes all of the mistakes in the dataset and replaces them with the corrected versions for grouping
    dataframe.replace({'County':{"BUENA VIST": "BUENA VISTA", "CERRO GORD": "CERRO GORDO", "OBRIEN": "O'BRIEN", "POTTAWATTA": "POTTAWATTAMIE"}}, inplace = True)
    #creates the county grouping and sums by county for printing
    countygroup = dataframe.groupby('County').sum()
    #prints out the sales dollars with formatting
    print1 = (countygroup["Sale (Dollars)"]).apply(lambda x: "${:,.2f}".format(x))
    print(print1)
    #prints out bottles sold, volume sold, and sale dollars by county with formatting
    print2 = (countygroup["Bottles Sold"]).apply(lambda x: "{:,}".format(x)),(countygroup["Volume Sold (Gallons)"]).apply(lambda x: "{:,.2f}".format(x)),(countygroup["Sale (Dollars)"]).apply(lambda x: "${:,.2f}".format(x))
    print(print2)
#function call
dataset_info(dataframe)
```
